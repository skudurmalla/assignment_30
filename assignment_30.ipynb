Problem Statement

In this assignment students have to compress racoon grey scale image into 5 clusters. In
the end, visualize both raw and compressed image and look for quality difference.
The raw image is available in spicy.misc package with the name face.

```python
import numpy as np
from sklearn import cluster, datasets
import scipy.misc
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Visualize the gray scale image
im = scipy.misc.face(gray=True)
plt.figure(figsize=(11, 3.5))
plt.imshow(im, cmap=plt.cm.gray)
plt.show()

# Compress the gray scale image into 5 clusters
rows = im.shape[0]
columns = im.shape[1]
print(rows,columns)

image = im.reshape(rows*columns,1)
kmeans = KMeans(n_clusters = 5)
kmeans.fit(image)

clusters = np.asarray(kmeans.cluster_centers_) 
labels = np.asarray(kmeans.labels_)  
labels = labels.reshape(rows,columns); 
plt.imsave('compressed_racoon.png',labels);

# Visualize the compressed image
image = plt.imread('compressed_racoon.png')
plt.figure(figsize=(10, 3.6))
plt.imshow(image)
plt.show()
```
